# INDEX OF ACCESS TO COOKING SOLUTIONS

The Index of Access to Cooking Solutions shows the MTF tier ranking. The lowest tier among all the attributes determines the householdâ€™s overall tier ranking for access to cooking solutions.

```python
HIT_PATH = '../../../../src/'
institution_id = 7
lang = 'en'
import os,sys, folium
sys.path.insert(0, os.path.normpath(os.path.join(os.path.abspath(''), HIT_PATH)))
import hedera_types as hedera
import odk_interface as odk

mfi = hedera.mfi(institution_id,setPathBook=True)
data = mfi.read_survey(mfi.odk_data_name)
mfi.HH = odk.households(data)
attribute = 'C_Index'
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
mfi.tier_barh([attribute],names_labels = ['Access to Cooking Solutions'])
```
