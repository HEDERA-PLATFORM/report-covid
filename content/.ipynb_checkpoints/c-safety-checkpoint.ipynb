# SAFETY

For health & safety, there are two categories: Tiers 3 & 5. If households have had cooking related accidents that required medical assistance in the past year and/or do not have sufficient ventilation in their kitchens, they are put in Tier 3. Accident-free households with adequately ventilated kitchens are ranked in Tier 5. 

```python
HIT_PATH = '../../../../src/'
institution_id = 7
lang = 'en'
import os,sys, folium
sys.path.insert(0, os.path.normpath(os.path.join(os.path.abspath(''), HIT_PATH)))
import hedera_types as hedera
import odk_interface as odk

mfi = hedera.mfi(institution_id,setPathBook=True)
data = mfi.read_survey(mfi.odk_data_name)
mfi.HH = odk.households(data)
attribute = 'C_Safety'
```
