# AVAILABILITY

Another indicator examined in the study was the availability of cooking fuel throughout the year. There are 3 possible tiers: 3, 4, & 5. In Tier 4, fuel is available at least 80% of the year, whereas it is available less than 80% in Tier 3. In Tier 5, it is always available.

```python
HIT_PATH = '../../../../src/'
institution_id = 7
lang = 'en'
import os,sys, folium
sys.path.insert(0, os.path.normpath(os.path.join(os.path.abspath(''), HIT_PATH)))
import hedera_types as hedera
import odk_interface as odk

mfi = hedera.mfi(institution_id,setPathBook=True)
data = mfi.read_survey(mfi.odk_data_name)
mfi.HH = odk.households(data)
attribute = 'C_Availability'
```

## MTF Ranking
